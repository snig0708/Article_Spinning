```python
import numpy as np
import random # for random number generation
import nltk
from bs4 import BeautifulSoup #XML parsing
```

```python
## Importing the data: Positive reviews from Amazon

positive_reviews = BeautifulSoup(open("/Users/snigdhacheekoty/Documents/GitHub/Article_Spinning/positive.review.rtf").read())
positive_reviews = positive_reviews.findAll("review_text")
```

# Collecting all trigrams

```python


trigrams = {}  # Dictionary to store the trigrams

for review in positive_reviews:
    s = review.text.lower() # converting the sentences to lowercase
    tokens = nltk.tokenize.word_tokenize(s) # tokenizing the sentence
    

# Fetching all possible trigrams from the sentences

    for i in range(len(tokens)-2): # counting from zero to the last third word
        k = (tokens[i], tokens[i+2]) # Creating the key as a tuple of (i, i+2)th words---> first and third words
                                     # we create a tuple because, it's a tuple: immutable and assign it to keys
        
        if k not in trigrams:
            trigrams[k]= [] # initialize the key in the array
        trigrams[k].append(tokens[i+1])

```

```python
## Printing the trigrams

print(trigrams)
```

# Converting these trigrams into probabilities

```python
trigrams_probabilities = {}
for k,words in trigrams.items():
    if len(set(words))>1:  # We need only those lists that have more than one middle word 
                           # because: if there's only one middle word, the word is replaced by the same word
        
        d ={} # this dictionary is keyed by the middle words.... to store values of counts against them
        n = 0 # Stores the number of middle words (for each key) within the list
        for w in words:
            if w not in d:
                d[w] = 0    # count given zero if word not found in the list
            d[w] += 1       # count incremented by 1 if word found in the list
            n += 1
        for w,c in d.items(): # Converting the count into probabilities
            d[w] = float(c)/n
        trigrams_probabilities[k] = d #replacing the list of words with dictionary of probabilties against the words

     
    
```

# Creating a function for randomly sample trigrams

```python
def random_sample(d):
    r = random.random() # generating a random number
    cumulative_probability = 0
    for w,p in d.items():
        cumulative_probability += p
        if r < cumulative_probability:
            return w
        
```

# Create a function to test the spinner

```python
def test_spinner():
    review = random.choice(positive_reviews)
    s = review.text.lower()
    print("Original:", s)
    tokens = nltk.tokenize.word_tokenize(s)
    for i in range(len(tokens) - 2):
        if random.random() < 0.2: # 20% chance of replacement
            k = (tokens[i], tokens[i+2])
            if k in trigrams_probabilities:
                w = random_sample(trigrams_probabilities[k])
                tokens[i+1] = w
    print("Spun:")
    print(" ".join(tokens).replace(" .", ".").replace(" '", "'").replace(" ,", ",").replace("$ ", "$").replace(" !", "!"))
    
```
